Api to model

```python
# main.py
from fastapi import FastAPI
from pydantic import BaseModel
import joblib
import pandas as pd

# تحميل النموذج
model = joblib.load("laptop_price_model.pkl")

app = FastAPI(title="Laptop Price Prediction API")

# Schema للـ input
class Laptop(BaseModel):
    Company: int
    TypeName: int
    Inches: float
    Ram: int
    OpSys: int
    Weight: float
    X_ScreenResolution: int
    Y_ScreenResolution: int
    Cpu_brand: int  
    Gpu_brand: int
    Memory_size: int

@app.post("/predict")
def predict_price(laptop: Laptop):
    # تحويل البيانات DataFrame
    df = pd.DataFrame([laptop.dict()])
    prediction = model.predict(df)[0]
    return {"Predicted Price": prediction}
```
